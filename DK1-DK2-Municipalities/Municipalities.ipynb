```python
import pandas as pd
```

```python
kommuner = '/content/drive/MyDrive/Colab Notebooks/Energinet/Municipalities/municipalities.csv'

kommuner = pd.read_csv(kommuner, index_col = 0)

kommuner.head()
```

```python
def area(row):
  if row['Region'] == 'Hovedstaden':
    val = 'DK2'
  elif row['Region'] == 'Sjælland':
    val = 'DK2'
  else:
    val = 'DK1'
  return val
```

```python
kommuner['Area'] = kommuner.apply(area, axis = 1)

kommuner.head()
```

```python
koder = '/content/drive/MyDrive/Colab Notebooks/Energinet/Municipalities/municipality_codes.xlsx'

koder = pd.read_excel(koder, converters = {'Kode':str, 'Navn':str})

koder.head()
```

```python
df = pd.merge(kommuner, koder, on = 'Navn', how = 'left')

df = df[['Kode', 'Navn', 'Region', 'Area']]

df.head()
```

```python
dk_1 = df[df['Area'] == 'DK1']
dk_2 = df[df['Area'] == 'DK2']

list_dk_1 = dk_1['Kode'].tolist()
list_dk_2 = dk_2['Kode'].tolist()

len(list_dk_1), len(list_dk_2)
```
