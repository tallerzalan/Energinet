https://www.energidataservice.dk/tso-electricity/Elspotprices#metadata-info

```python
import requests
```

```python
response = requests.get(
    url='https://api.energidataservice.dk/dataset/Elspotprices?offset=0&start=2022-10-01T00:00&end=2022-10-02T00:00&sort=HourUTC%20ASC&timezone=utc')

result = response.json()

records = result.get('records', [])

for record in records:
  print(' ', record)
```
